Ride Sharing Data Analysis

```python
import numpy as np

import pandas as pd
```

```python
df = pd.read_csv(r"D:\python book DHP\uber.csv")
df
```

1)Calculate the average fare amount for all rides

```python
avg_fare = df["fare_amount"].mean()
print("avg_fare amount is:",avg_fare)
```

2)Find the top 10 Rides with the highest fare amount.

```python
top10_fare = df.sort_values(by='fare_amount', ascending=False)["Unnamed: 0"].head(10) #Unnamed:0 we show according to all columns
print(top10_fare)
```

3)Calculate the total Revenue Generated from all rides.

```python
total_revenue = df['fare_amount'].sum()
print("Total revenue generated from all rides:",total_revenue)
```

4)Determine the average fare amount for each passenger count

```python
avg_passenger = df.groupby('passenger_count')['fare_amount'].mean()
avg_passenger
```

5)Find the  maximum and minimum fare amount.

```python
max_fare = df['fare_amount'].max()
min_fare = df['fare_amount'].min()

print("Maximum fare amount:",max_fare)
print("minimum fare amount:",min_fare)
```

6)Extract hour from pickup_datetime and identity the number of rides per hour.

```python

df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
df['hour'] = df['pickup_datetime'].dt.hour

df['hour'].value_counts().sort_index()
```

7)Determine the average fare amount for each hour of the day.

```python
avg_hour_fare = df.groupby('hour')['fare_amount'].mean()
print("The average fare amount for each hour of the day:",avg_hour_fare)
```

8)Find the total number of rides for each passenger count.

```python
rides_passenger = df['passenger_count'].value_counts().sort_index()
rides_passenger
```

9)Identify the top 10 rides with the highest passenger count.

```python
top10_passenger = df.sort_values(by='passenger_count', ascending=False).head(10)

print("Top 10 rides with the highest fare amount:",top10_passenger)
```

10)Calculate the daily number of rides using pickup_datetime.

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])

daily_rides = df.groupby(df['pickup_datetime'].dt.date).size()

print(daily_rides)
```
